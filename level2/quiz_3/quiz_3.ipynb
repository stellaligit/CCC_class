## Chicken and Rabbit
"Chicken and rabbit" is a famous math problem. As well known, each chicken has one head and two legs, while each rabbit has one head and four legs. Given the statement that the total number of heads for both chicken and rabbit is X and the total number of legs is Y. Please write a program to determin if there is a solution to satisfy the statement.

### Input Specification
The first line of input will contain 2 integers, X and Y, (1≤X,Y≤100), the total number of head and the total number of legs.

### Output Specification
Output Yes, if there is a solution to satisfy the given statement; otherwise, output No.

### Constraints
Subtask	| Points | Additional constraints <br>
1 | 40 | X ≤ 3 <br>
2 | 60 | No additional constraints. <br>

### Sample Input 1
4 8
### Sample Output 1
Yes
### Explanation
If there are 4 chickens, they have 4 heads and 8 legs.

### Sample Input 2
3 8
### Sample Output 2
Yes
### Explanation
If there are 2 chickens and 1 rabbit, they have 3 heads and 8 legs.

### Sample Input 3
1 8
### Sample Output 3
No

### Bob's Hated Number List
Bob has a hated number list, which has N numbers, denoted as a1,a2,…,aN. Bob tries to find a number nearest to the target number X, but he cannot use any number in his hated number list. Can you help Bob find out an integer Y which is not in the hated number list and has the min absolute difference with X. If there are multiple such integers, find out the smaller one.

### Input Specification
The first line of input contains two integers, X and N (1≤X≤106, 0≤N≤100), the target number and the length of the hated number list.

The second line of input contains N integers, ai, (1≤ai≤106), a number Bob hates.

### Output Specification
Output one integer, which is not in the hated number list and has the min absolute difference with X.

### Constraints
Subtask	Points | Additional constraints <br>
1 | 30 | N ≤ 5 <br>
2 | 70 | No additional constraints. <br>
### Sample Input 1
5 4 <br>
2 3 4 5 <br>
### Sample Output 1
6
### Sample Input 2
10 5 <br>
4 7 10 6 5 <br>
### Sample Output 2
9
### Sample Input 3
3 5
7 8 9 10 11
### Sample Output 3
3

## Bob's Word Conversion
Bob has two words A and B. Bob wants to know if it is possible to conver the word A to the word B by removing at most one character from the word A. Can you write a program to help him?

Input Specification
The first line of input contains the word A, while the second line of input contains the word B (1≤|A|,|B|≤105), where both A and B are strings consisting of lower case English letters.

Output Specification
If Bob can convert the word A to the word B by removing at most one character from A, output Yes; otherwise, output No.

### Constraints
Subtask | Points | Additional constraints <br>
1 | 20 | |A|, |B| ≤ 10 <br>
2 | 20 | |A|, |B|≤1000. <br>
3 | 60 | No additional constraints. <br>

### Sample Input 1
world
word
### Sample Output 1
Yes
### Explanation:
Bob can remove the letter l to convert world to word

### Sample Input 2
snekee
snuke
### Sample Output 2
No
### Sample Input 3
hello
hello
### Sample Output 3
Yes
