## Doubling or Losing

Alice and Bob are playing games again. If Bob gives Alice n dollars, Alice will give Bob double the amount. However, Alice is not good at math. When she calculates the result of 2 x n, sometime she will forget the last digit in the result. For example, if Bob gives Alice 12 dollars, the double amount should be 2 x 12 = 24. But Alice may get the incorrect result of 2 by missing the last digit, and thus gives back Bob  dollars. Now, Bob wants to know if he gives  dollars to Alice, how much money he may lose in the worst case.

### Input Specification
The first line of input contains one integer, n (5 ≤ n ≤ 10<sup>15</sup>), the amount of money Bob will give to Alice.

### Output Specification
Output one single integer, the mount of money Bob may lose in the worst case.

### Constraints
Subtask | Points | Additional constraints <br>
1 | 30 | n ≤ 10<sup>8</sup>  <br>
2 | 7 | No additional constraints  <br>

### Sample Input 1
12
### Sample Output 1
10
### Explanation
When calculating 2 x 12 = 24, Alice may get the incorrect result of 2. Thus, Bob will lose 12 - 2 = 10 dollars.

### Sample Input 2
3502
### Sample Output 2
2802
### Explanation
When calculating 2 x 3502, Alice will get the incorrect result of 700. Thus, Bob will lose 3502 - 700 = 2802 dollars.
