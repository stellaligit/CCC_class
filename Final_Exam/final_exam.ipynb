## Who is the first?
 n people stand in a line, numbered from 1 to n from front to back. They will walk through  doors. Each time they reach a door, the first person in the line will hold the door open, and everyone after the first person will walk through the door in order. After everyone is through the door, the person who holdes the door will join the line in the back.

Find out the first person in the line after walking through m doors.

### Input Specification
The first line of input contains two integers n and m (1 ≤ n, m ≤ 10 ^ 18), indicating the number of people and the number of doors.

### Output Specification
Ouput a single integer, the first person in the line after walking through m doors.

### Sample Input 1
5 2

### Sample Output 1
3

### Sample Input 2
4 101

### Sample Output 2
2

6 3 4 <br>

5

1 <br>

## Robot Control
Bob is attending a robot competition. The competition ground can be modelled as a number line. There are M lights, located at coordinates p1, p2, ... pM. The target of the competition is to turn on all lights.

Bob has N robots and he can place each robot at some integer location initially. It's also possible to place multiple robots at the same coordinate. Once the competition starts, Bob will control his robot to move. In one move, Bob can choose one robot at coordinate x and move it to x + 1 or x - 1. When a robot reaches a light's location, the light will be turned on. Bob wants to find the minimum number of moves he needs to turn on all lights. Can you write a program to help him?

### Input Specification
The first line of input contains two integers N and M, the number of robots and the number of lights.

The second line of input contains M integers, Pi, the coordinate of the i-th light. It guarantees that all pi are distinct.

### Output Specification
Output one integer, the minimum number of moves Bob needs to turn on all lights.

### Sample Input 1
3 5 <br>
10 20 3 1 18 <br>
### Sample Output 1
4
### Explanation
Bob can place the 1st robot at coordinate 3, the 2nd robot at coordinate 10, and the 3rd robot at coordinate 18. Then, Bob can take two moves to move the 1st robot to coordinate 1 and two moves to move the 3rd robot to coordinate 20. It takes four moves in total.

### Sample Input 2
3 3 <br>
-100 2000 53 <br>
### Sample Output 2
0

## Tree Coloring
Given a tree with n nodes and n-1 edges, the i-th edge connects the vertices ui and vi with a length of wi. Bob wants to color each node either red or black so that any path connecting two nodes with the same color has an even distance. Can you write a program to help Bob color the tree?

### Input Specification
The first line of input contains an integer n, the number of nodes in the tree.

Each of the following n-1 lines contains three integers ui, vi and wi, indicating an edge connecting ui and vi with a length of wi.

### Output Specification
Output n integers in one line. The i-th number is 0 if node i is colored in red and 1 if it's colored in black. If there are multiple solutions, any of them will be accepted.

### Sample Input 1
3 <br>
1 2 2 <br>
2 3 1 <br>
### Sample Output 1
1 1 0
### Sample Input 2
5 <br>
2 5 2 <br>
2 3 10 <br>
1 3 8 <br>
3 4 2 <br>
### Sample Output 2
1 0 1 0 1
### Explanation
There are some other possible solutions, like [1, 1, 1, 1, 1]. You just ned to output any one of the valid solutions.

### Landscaping

Bob has a backyard which can be modelled as a M by N rectangular grid with M rows and N columns. He wants to plant some flowers in some cells. Specially, some of these cells are occupied by obstacles and cannot be planted. Also, Bob knows that two flowers cannot be too close; otherwise they will compete for sunlight, water and nutrients. So, Bob will never choose two cells if they share an edge.

Now, Bob wants to find out the number of ways he can choose some cells to plan flowers. It's also possible that Bob doesn't choose any cell. Since the answer is huge, output the answer modulo 10 ^ 9 + 7.

### Input Specification
The first line of input contains two integers M and N, the number of rows and the number of columns.

Each of the following M lines contains N integers, ai,j, where 0 indicates obstacle while 1 indicates that Bob can plant flower at that cell.

### Output Specification
Output one integer, the number of ways for Bob to plant flowers modulo 10 ^ 9 + 7.

### Sample Input
2 3 <br>
1 1 1 <br>
0 1 0 <br>
### Sample Output
9
### Explanation
There is 1 way for Bob to plant 0 flowers, 4 ways to plant 1 flower, 3 ways to plant 2 flowers, and 1 way to plant 3 flowers. Thus, the total number of ways is 1 + 4 + 3 + 1 = 9.
